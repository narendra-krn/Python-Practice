```python
x = np.arange(1,17)
print(x)
print(x.ndim)
print(x.shape)
```

```python
n = x.reshape((4,4))
print(n)
print(n.shape)
print(n.ndim)
```

```python
n1 = x.reshape((4,2,2))
print(n1)
print(n1.shape)
print(n1.ndim)
```

```python
v=n.ravel() #here we use ravel it will convert the array into 1 dimensional array(if the array was 2D or 3D)
print(v)
print(v.shape)
print(v.ndim)
```

```python
v=n.flatten() #here we use flatten it will convert the array into 1 dimensional array(if the array was 2D or 3D)
print(v)
print(v.shape)
print(v.ndim)
```
