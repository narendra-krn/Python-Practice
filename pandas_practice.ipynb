```python
#get_dummies() in pandas is used to convert categorical data (text categories) into numeric columns.
#This process is called One-Hot Encoding, which is commonly used in machine learning and data preprocessing.

#Many ML algorithms cannot understand text values like "Male", "Female", "Food", "Travel". So get_dummies() converts them into 0 and 1 columns.

import pandas as pd

data = {
    "firstname": ["Raja", "Anu", "Ravi", "Deepu"],
    "lastname": ["narendra", "rani", "raja", "singh"],
    "employmenttype": ['Service', 'Business', 'Student', 'Service'],
    "country": ['Australia', 'usa', 'sweden', 'india']
}
df = pd.DataFrame(data, columns= ['firstname', 'lastname', 'employmenttype', 'country'])
print(df)
```

```python
df1 = pd.get_dummies(df['employmenttype'], dtype=int)
df2 = pd.get_dummies(df['country'], dtype=int)
print(df1)
print(df2)
```

```python
import pandas as pd
india_weather = pd.DataFrame({
    "city" : ['vizag', 'bangalore', 'hyderabad'],
    "temprature" : [40, 29, 37],
    "humidity": [80, 40,60]
})
india_weather
```

```python
import pandas as pd
us_weather = pd.DataFrame({
    "city" : ['new york', 'chicago', 'orlando'],
    "temprature" : [40, 29, 37],
    "humidity": [80, 40,60]
})
us_weather
```

```python
df= pd.concat([india_weather, us_weather])
df
```

```python
df= pd.concat([india_weather, us_weather], ignore_index=True)
df
```
